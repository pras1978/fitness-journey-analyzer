Step 2 — Create streamlit_app.py

```python
code = '''
import streamlit as st
import pandas as pd
import os

st.set_page_config(page_title="Fitness Journey Analyzer", layout="wide")

st.title("🏋️ Intelligent Fitness Journey Analyzer")

BASE_PATH = "/content/drive/MyDrive/fitness-journey-analyzer/data/processed"

# ---------------------------
# Load files
# ---------------------------

pose_file = os.path.join(BASE_PATH, "pose_scores_sample.csv")
nlp_file = os.path.join(BASE_PATH, "nlp_workout_log_results.csv")
forecast_file = os.path.join(BASE_PATH, "fitness_forecast.csv")

pose_df = pd.read_csv(pose_file)
nlp_df = pd.read_csv(nlp_file)
forecast_df = pd.read_csv(forecast_file)

# ---------------------------
# Dashboard Layout
# ---------------------------

col1, col2 = st.columns(2)

with col1:
    st.subheader("Posture Scores")
    st.dataframe(pose_df)

with col2:
    st.subheader("Workout Sentiment")
    st.dataframe(nlp_df)

# ---------------------------
# Forecast Section
# ---------------------------

st.subheader("Fitness Progress Forecast")

st.line_chart(forecast_df["yhat"])

# ---------------------------
# Insight Section
# ---------------------------

st.subheader("AI Insights")

latest_posture = pose_df["posture_score"].mean()
sentiment = nlp_df["sentiment"].mode()[0]

st.write("Average Posture Score:", round(latest_posture,2))
st.write("Most common sentiment:", sentiment)

if latest_posture < 0.6:
    st.warning("Posture needs improvement.")
else:
    st.success("Good posture detected!")

'''

path = "/content/drive/MyDrive/fitness-journey-analyzer/frontend/streamlit_app.py"

with open(path,"w") as f:
    f.write(code)

print("streamlit_app.py created")
```
