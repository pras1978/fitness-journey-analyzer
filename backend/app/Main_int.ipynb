Step 3 — Create main.py

```python
code = '''
from fastapi import FastAPI
import pandas as pd
import os

app = FastAPI(title="Intelligent Fitness Journey Analyzer")

BASE_PATH = "/content/drive/MyDrive/fitness-journey-analyzer/data/processed"

@app.get("/")
def home():
    return {"message": "Fitness Journey Analyzer API running"}

@app.get("/dashboard")
def dashboard():
    file = os.path.join(BASE_PATH, "final_dashboard_summary.csv")
    df = pd.read_csv(file)
    return df.to_dict(orient="records")

@app.get("/forecast")
def forecast():
    file = os.path.join(BASE_PATH, "fitness_forecast.csv")
    df = pd.read_csv(file)
    return df.tail(30).to_dict(orient="records")
'''

path = "/content/drive/MyDrive/fitness-journey-analyzer/backend/app/main.py"

with open(path, "w") as f:
    f.write(code)

print("main.py created at:", path)
```
